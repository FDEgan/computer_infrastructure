**Background of the Dataset**

The Met Éireann WDB API outputs a detailed point forecast in XML format for a coordinate point as defined by the user.

Data is output in (i) 1-hour intervals out to 90 hours, (ii) three-hour intervals from 90h to 144 hours, and (iii) six-hour intervals from 144 hours out to 240 hours.

National Coverage The Met Éireann WDB API outputs a detailed point forecast in XML format for a coordinate point as defined by the user. Data is output in (i) 1-hour intervals out to 90 hours, (ii) three-hour intervals from 90h to 144 hours, and (iii) six-hour intervals from 144 hours out to 240 hours.The meteorological parameters for each timestep are as follows: Timing, Location, Temperature, Wind Direction and speed, Humidity, Pressure, Cloud, Dewpoint, Temperature, Precipitation, Weather symbol This API provides the data used by the point forecasts on Met Éireann's main website and mobile apps.

**Overview of Variables**

**Categorical:**

- **`name`:** This variable indicates the name of the weather staition from which the dataset is derivd. It has one value: Athenry.
- **`symbol`:** This variable indicates whether its Day or Night (n/d) and the number that proceeds the symbol is the description of the weather e.g cloudy, fair, light rain.
- **`weatherDescription`:** This variable indicates the type of weather. It has three categories: cloudy, fair, light rain.
- **`text`:** This variable indicates the type of weather. It has three categories: cloudy, fair, Recent Drizzle.
- **`cardinalWindDirection`:** This variable indicates the direction of the wind. It has two categories: W (West), SW (South West).
- **`windDirection`:** This variable indicates the direction of the wind. It is the wind direction in in dgrees on a compass. It has two categories: 225, 270.
- **`dayName`:** This variable indicates day the weather data was gathered. It has one value: Monday.
- **`date`:** This variable indicates the date the data was gathered. It has one value: 2024-11-25.
- **`reportTime`:** This variable indicates the hour at which the data was gathered. It has 23 values, one for each hour from midnight:

These variables are categorical as: 

1. All the values cannot be measured.
2. Mathematical operations cannot be performed on them.
3. There is no way to rank them.
4. There are only a distinct number of values in the variable.


**Numerical:**

- **`temperature`:** The degrees celcius when the data was gathered.
- **`windSpeed`:** The speed of the wind measured in kmh.
- **`windGust`:** The speed of the wind gust (a sudden, brief increase in wind speed) measured in kmh.
- **`humidity`:** The level of humidity in the air represented as a percentage.
- **`rainfall`:** The precipitation in millimetres.
- **`pressure`:** Hectopascals measurment from a barometer.


These variables are numerical as:

1. All the values can be measured.
2. Numerical operations can be performed on them.
3. They can be ordered or ranked based on value.
4. There is a infinite number of values.


```python
df['reportTime'] = pd.to_datetime(df['reportTime']) # Converting the reportTime field to be datetime

# Plot temperature over time
plt.figure(figsize=(10, 6)) # Designating the size of the plot
plt.plot(df['reportTime'], df['temperature'], marker='o', linestyle='-', color='b') # Stating the fields to be used and the adding some formatting. 

plt.title('Temperature Over Time', fontsize=16) # Title for the plot.
plt.xlabel('Report Time', fontsize=12) # Title for the X axis
plt.ylabel('Temperature (°C)', fontsize=12) # Title for the Y axis

plt.xticks(rotation=45) # To improve readability, slightly tilting the x-labels.

plt.grid(False) # Removing gridlines as I feel its cleaner

plt.tight_layout()  # Ensuring labels don't overlap.
plt.show() # Showing the plot
```
